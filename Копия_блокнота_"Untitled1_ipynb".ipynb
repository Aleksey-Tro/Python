# 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
df = df_1.groupby('price')['yr_built']

plt.figure(figsize=(8, 6))
plt.scatter(df_1['yr_built'], df_1['price'])

plt.title('Зависимость цены от года построки')
plt.xlabel('Год постройки')
plt.ylabel('Цена');
```

Данная диаграмма показывает тенденцию понижения цены на более старую нидвижемость. Малая часть старой недвижимости лишь повышает свой price, скорее всего она входит в редкую категорию недвижимости, которая с годами только дороже(исторически важная недвижимость, например: дом где родилсяили или жил известный человек, замок и т.д.).

```python
df = df_1.groupby('price')['condition'].mean().value_counts().reset_index()

plt.figure(figsize=(8, 6))
plt.bar(df_1['condition'], df_1['price'])

plt.title('Зависимость цены от состояния недвижимости')
plt.xlabel('Cостояния недвижимости')
plt.ylabel('Цена');
```

На диограмме видно повышение цены недвижимости в среднем и хорошем состоянии. Удивление вызывает низкая ценовая категория недвижимости в отличном состоянии. Многие факторы могли послужить этому, местонахождение, год постройки, вид, планировка. Мое предположение, что такой низкий price на недвижимость в хорошем состоянии оказывает планировка или спрос потребителя, т.к. недвижимость в хорошем состоянии за частую является недавно построенной и цена должна быть высокой, но нужды потребителей могут привести к резкому снижению спроса, а соответственно  цены за счет слабой экономики, менталитета(возможно они привыкли жить в старых или фамильных домах) или чего-то другого. На первый взгляд трудно сказать в чем причина таких данных, необходимо делать полный анализ ситуации.

```python
df = df_1[['price', 'sqft_living', 'grade']].corr()
df = np.round(df, 1)
df[np.abs(df) < 0.3] = 0

plt.figure(figsize=(8,6))
sns.heatmap(df, annot=True, linewidths=.5, cmap='coolwarm')
```

Матрица корреляций показывает зависимость цены от жилой площади и качества жилья.
